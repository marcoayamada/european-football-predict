# Discussão sobre features

Algumas colunas podem ser consideradas como componentes para formar uma Time serie. Julguei `round`, `time` e `home/away_score`. 

As 2 primeiras ok, mas porque a última? O mesmo time fabrica um histórico durante o campeonato e durante os anos isso pode se tornar uma informação muito importante. É comum noticiar que o time x não ganha do time y há tanto tempo, principalmente em clássicos.

```python
# convertendo o nome dos times para numeros

"""
Nesse caso vamos usar o LabelEncoder do sklearn pois existem centenas de times.
Essa transformação é necessária para que possamos fazer isso com novos dados.
"""

dft = df[['home_name', 'away_name']]
le = preprocessing.LabelEncoder()

name_trans = dft.apply(le.fit_transform)

# para fazer a conversao inversa. Agora nao precisarei, mas depois sim.
# name_trans.apply(le.inverse_transform)

# substituindo os valores
df.home_name = name_trans.home_name
df.away_name = name_trans.away_name
```

### 2) Removendo tempo, rodada

O score Elo leva em consideração o último jogo do time. Dependendo do resultado esse valor é alterado. Então
desconsiderei o tempo pois ele já está implicito no score Elo.

Esse conjunto de dados é uma Time series, porém o modelo poderia ficar muito complexo devido à variedade de ligas e os dias
que acontecem os jogos.

Sabemos que uma rodada ao final do campeonato com o time já campeão pode significar uma derrota para um time fraco, porém
esse é um comportamento anormal e prever isso poderia também tornar o modelo complexo. Para esses casos o feeling do apostador
vale mais do que qualquer previsão.


```python
# removendo a coluna tempo e rodada

df.drop(['time', 'round'], axis=1, inplace=True)
```

```python
df.head(10)
```

### 3) last5all: Juntando 3 colunas em uma

Existem 3 colunas com essa característica: win, draw e lose. Juntando essas 3 colunas em uma calculando o aproveitamento de pontos.

```python
def aprov_last5_home(col):
    aprov_win = col['home_last5all_home_win']*3
    aprov_draw = col['home_last5all_home_draw']*1
    
    return int((aprov_win+aprov_draw)/15*100)

def aprov_last5_away(col):
    aprov_win = col['away_last5all_away_win']*3
    aprov_draw = col['away_last5all_away_draw']*1
    
    return int((aprov_win+aprov_draw)/15*100)
```

```python
df['aprov_5_home'] = df[['home_last5all_home_win', 'home_last5all_home_draw']].apply(lambda x: aprov_last5_home(x), axis=1)
df['aprov_5_away'] = df[['away_last5all_away_win', 'away_last5all_away_draw']].apply(lambda x: aprov_last5_away(x), axis=1)
```

```python
df.drop(['home_last5all_home_win', 'home_last5all_home_draw', 'home_last5all_home_lose',
         'away_last5all_away_win', 'away_last5all_away_draw', 'away_last5all_away_lose'], axis=1, inplace=True)
```

```python
# Função que monta um dataframe para exibir as feature importances
def view_feature_importances(feat, impor):
    return pd.DataFrame({'features': feat, 'importances': impor}).sort_values(by='importances', ascending=False).set_index('features')
```

```python
view_feature_importances(x.columns.values, clf.feature_importances_)
```

Como mencionei no começo, o resultado do jogo - `home/away_score` tem uma importancia grande na previsão - 76%. Não podemos deixar o modelo dessa forma, pois não sabemos qual será o resultado do jogo. Mas antes de remover completamente essa feature vamos testar para ver como a previsão fica setando todos os jogos em 0x0:

```python
x_train.home_score = 0
x_train.away_score = 0
```

```python
clf2 = RandomForestClassifier(n_estimators=100, n_jobs=-1, max_depth=8, min_samples_leaf=50, random_state=99, verbose=0)
clf2.fit(x_train, y_train)
```

```python
accuracy_score(y_test, clf2.predict(x_test))
```

```python
view_feature_importances(x_train.columns.values, clf2.feature_importances_)
```

Agora sim. O modelo parece estar mais próximo da realidade. Pode-se notar que o resultado teve importância 0 para compor o modelo. Outra observação importante é o score que agora está em 0.5, o que era esperado já que não foi feita nenhuma mudança nas features.

Agora precisamos trabalhar em cima desses valores e fazer uma engenharia das features:

```python
y = df['final_result']
x = df.drop('final_result',axis=1)

# removendo o resultado da partida
x.drop(['home_score', 'away_score'], axis=1, inplace=True)
```

```python
x.head(8)
```

As escalas dos valores são muito variadas.... Por exemplo `away_last5all_away` vai de -17 a 28 `fifa_home_ova` vai de 0 a 100, `elo_score` vai mais ou menos de 1500 a 2000. Isso pode ser um problema para fazer o scalling das features. Nesse caso o mais indicado é usar a `Normalizer` que deixa todos os pontos a 1 de distância da origem. As colunas que constam os nomes dos times são dados categóricos mas estão representados por números. Não há problema em incluí-los na escala.

```python
# Função que calcula a feature importance

def calcula_fi(x_cols):
    norm = preprocessing.Normalizer()
    x_norm = norm.fit_transform(x[x_cols])
    x_train, x_test, y_train, y_test = train_test_split(x_norm, y, test_size=0.33, random_state=29384)
    clf3 = RandomForestClassifier(n_estimators=200, n_jobs=-1, max_depth=7, min_samples_leaf=50, random_state=99, verbose=0)
    clf3.fit(x_train, y_train)
    print('Accuracy: {0}'.format(accuracy_score(y_test, clf3.predict(x_test))))
    
    return view_feature_importances(x_cols, clf3.feature_importances_)
```

```python
x_cols =['home_name', 'away_name',
         'home_pos', 'away_pos',
         'home_last5all_home', 'away_last5all_away', 'last5all_home_away_dif',
         'fifa_home_ova', #'fifa_home_att', 'fifa_home_mid', 'fifa_home_def',
         'fifa_away_ova', #'fifa_away_att', 'fifa_away_mid', 'fifa_away_def',
         'elo_home_score', 'elo_away_score',
         'tfm_value_home', 'tfm_value_away',
         'aprov_5_home', 'aprov_5_away'
        ]

calcula_fi(x_cols).plot(kind='barh')
plt.show()
```

```python
x_cols =['home_name', 'away_name',
         #'home_pos', 'away_pos',
         #'home_last5all_home', 'away_last5all_away', 'last5all_home_away_dif',
         'fifa_home_ova', #'fifa_home_att', 'fifa_home_mid', 'fifa_home_def',
         'fifa_away_ova', #'fifa_away_att', 'fifa_away_mid', 'fifa_away_def',
         'elo_home_score', 'elo_away_score',
         'tfm_value_home', 'tfm_value_away',
         #'aprov_5_home', 'aprov_5_away'
        ]

calcula_fi(x_cols).plot(kind='barh')
plt.show()
```

```python
x_cols =['home_name', 'away_name',
         #'home_pos', 'away_pos',
         #'home_last5all_home', 'away_last5all_away', 'last5all_home_away_dif',
         #'fifa_home_ova', #'fifa_home_att', 'fifa_home_mid', 'fifa_home_def',
         #'fifa_away_ova', #'fifa_away_att', 'fifa_away_mid', 'fifa_away_def',
         'elo_home_score', 'elo_away_score',
         'tfm_value_home', 'tfm_value_away',
         #'aprov_5_home', 'aprov_5_away'
        ]

calcula_fi(x_cols).plot(kind='barh')
plt.show()
```

Não podemos confiar 100% no modelo, há outras questões que devem ser observadas.

Seguindo em frente, vamos pegar essas 6 features testar com outros algoritmos. Para isso não vamos separar elas em treino e teste, vamos usar cross-validation para aproveitar todos os dados e verificar se o modelo está em over/underfiting. Além disso vamos tentar fazer um tunning dos hyperparametros para ver se conseguimos alcançar um modelo melhor.

```python
def make_preds(x_cols, algs):
    norm = preprocessing.Normalizer()
    x_norm = norm.fit_transform(x[x_cols])
    
    for alg, name in algs:
        scores = cross_val_score(alg, x_norm, y, cv=10)
        print('{0} ~ {1}'.format(name, scores.mean()))
```

```python
make_preds(x_cols, algs)
```
